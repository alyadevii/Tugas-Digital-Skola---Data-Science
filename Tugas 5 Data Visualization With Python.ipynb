```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/BankChurners.csv')
display(data.head())
```

```python
data.Income_Category.unique()
```

```python
income_counts = data['Income_Category'].value_counts()
plt.barh(income_counts.index, income_counts.values, height=0.8, color = 'turquoise', edgecolor ='black')
plt.xlabel("Frequency")
plt.ylabel("Income Category")
plt.title("Bar Chart of Income Category")
plt.show()
```

Grafik batang horizontal tersebut menunjukkan distribusi kategori pendapatan berdasarkan frekuensinya. Dari visualisasi tersebut, terlihat bahwa kategori pendapatan "Less than 40K" memiliki jumlah tertinggi, yaitu sekitar 3.600, menunjukkan bahwa sebagian besar individu dalam data ini berada pada kelompok berpendapatan rendah. Kategori berikutnya secara frekuensi adalah "40K - 60K", "80K - 120K", dan "60K - 80K", meskipun urutan kategori di sumbu Y tidak tersusun secara numerik. Kategori "Unknown" menunjukkan jumlah individu yang tidak menyebutkan pendapatannya, dan kategori "$120K+" merupakan kelompok dengan jumlah terkecil, hanya sekitar 800. Grafik ini mengindikasikan ketimpangan distribusi pendapatan, dengan dominasi kelompok berpendapatan rendah dan sangat sedikit individu yang berada pada kelompok pendapatan tinggi.

```python
data.Education_Level.unique()
```

```python
education_counts = data['Education_Level'].value_counts()
plt.bar(education_counts.index, education_counts.values, width=0.8, color = 'palevioletred', edgecolor ='black')
plt.xlabel("Frequency")
plt.ylabel("Education Level Category")
plt.title("Bar Chart of Education Level Category")
plt.xticks(rotation=45)
plt.show()
```

Grafik batang tersebut menunjukkan distribusi kategori tingkat pendidikan berdasarkan frekuensinya. Dari visualisasi ini, dapat dilihat bahwa kategori "Graduate" memiliki frekuensi tertinggi, yaitu lebih dari 3.000 orang, yang menunjukkan bahwa mayoritas individu dalam data ini telah menyelesaikan pendidikan tingkat sarjana. Kategori selanjutnya yang paling banyak adalah "High School", diikuti oleh "Unknown" atau jumlah individu yang tidak menyebutkan pendapatannya dan "Uneducated", yang masing-masing memiliki frekuensi sekitar 1.500. Sementara itu, kategori "College", "Post-Graduate", dan "Doctorate" memiliki frekuensi yang lebih rendah, dengan "Doctorate" menjadi yang paling sedikit, kurang dari 500 orang. Grafik ini mengindikasikan bahwa sebagian besar populasi dalam data memiliki latar belakang pendidikan menengah hingga sarjana, sedangkan tingkat pendidikan pascasarjana dan doktoral jauh lebih sedikit.

```python
data.Gender.unique()
```

```python
gender_counts = data['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['indianred', 'cornflowerblue'], wedgeprops={'width':0.4, 'edgecolor':'black'})
plt.title('Gender')
plt.show()
```

Perbedaan ini tidak terlalu besar, tetapi menunjukkan bahwa jumlah perempuan sedikit lebih banyak dibandingkan laki-laki dalam dataset yang dianalisis. Visualisasi seperti ini membantu memahami komposisi gender secara proporsional dalam bentuk yang mudah dicerna.

```python
sns.countplot(data=data, x='Income_Category', hue='Education_Level', palette='pastel')
plt.xlabel ("Income Category")
plt.ylabel ("Count")
plt.title ("Countplot of Income & Education")
plt.tight_layout
plt.xticks(rotation=45)
plt.show()
```

Grafik countplot tersebut menunjukkan hubungan antara kategori pendapatan (Income Category) dan tingkat pendidikan (Education Level). Terlihat bahwa sebagian besar individu berada dalam kategori pendapatan "Less than 40K", dengan mayoritas berasal dari tingkat pendidikan High School dan Graduate. Kategori pendapatan ini memiliki jumlah individu tertinggi dibandingkan kategori lainnya. Pada kelompok pendapatan menengah seperti "60K - 80K" dan "80K - 120K", jumlah individu juga cukup besar dan lebih merata dalam hal tingkat pendidikan, meskipun lulusan Graduate tetap mendominasi. Sementara itu, pada kategori pendapatan tinggi "$120K+", jumlah individu dari semua tingkat pendidikan relatif rendah, namun cenderung didominasi oleh individu dengan pendidikan Graduate dan Post-Graduate. Pendidikan Uneducated dan Unknown tersebar di semua kategori pendapatan, tetapi paling banyak terdapat di kelompok berpendapatan rendah. Grafik ini mencerminkan adanya kecenderungan bahwa semakin tinggi tingkat pendidikan seseorang, maka semakin besar kemungkinan mereka berada pada kelompok pendapatan yang lebih tinggi.

```python
Avg_Utilization_Ratio = data['Avg_Utilization_Ratio']
Total_Trans_Ct= data['Total_Trans_Ct']
plt.scatter(Avg_Utilization_Ratio, Total_Trans_Ct, color = 'cadetblue')
plt.xlabel ("Age")
plt.ylabel ("Dependent Count")
plt.title ("Scatter Plot of Age vs Dependent Count")
plt.show()
```

Grafik scatter plot ini menunjukkan hubungan antara usia (Age) dan jumlah tanggungan (Dependent Count). Namun, pada grafik ini sumbu usia (x-axis) terlihat sudah dinormalisasi atau diubah ke skala antara 0 hingga 1, sehingga tidak menunjukkan nilai usia asli secara langsung.

Dari plot tersebut, tidak terlihat pola yang jelas atau korelasi kuat antara usia dan jumlah tanggungan, karena titik-titik tersebar cukup merata di seluruh rentang nilai. Ini mengindikasikan bahwa dalam data tersebut, jumlah tanggungan tidak secara signifikan dipengaruhi oleh usia pelanggan.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

num_cols = ['Customer_Age', 'Dependent_count', 'Months_on_book', 'Total_Relationship_Count',
            'Months_Inactive_12_mon', 'Contacts_Count_12_mon', 'Credit_Limit',
            'Total_Revolving_Bal', 'Avg_Open_To_Buy', 'Total_Amt_Chng_Q4_Q1',
            'Total_Trans_Amt', 'Total_Trans_Ct', 'Total_Ct_Chng_Q4_Q1',
            'Avg_Utilization_Ratio', 'Dependent_count_jitter']

corr_matrix = data[num_cols].corr()

mask = np.triu(np.ones_like(corr_matrix, dtype=bool))


cmap = sns.diverging_palette(220, 20, as_cmap=True)

plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, mask=mask, cmap='viridis', annot=True, square=True,
            linewidths=0.1, vmax=1, vmin=-1, annot_kws={"size": 9, "color": "black"})
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```

Heatmap korelasi ini menunjukkan hubungan antara berbagai fitur numerik dalam dataset. Terlihat bahwa terdapat korelasi positif yang kuat antara usia pelanggan (Customer_Age) dengan lama keanggotaan dalam bulan (Months_on_book), yang berarti pelanggan yang lebih tua cenderung memiliki akun lebih lama. Selain itu, total jumlah transaksi (Total_Trans_Amt) sangat berkorelasi dengan jumlah transaksi (Total_Trans_Ct), yang masuk akal karena semakin banyak transaksi biasanya total jumlahnya juga lebih besar. Sementara itu, terdapat korelasi negatif sedang antara batas kredit yang tersedia untuk digunakan (Avg_Open_To_Buy) dengan saldo yang berputar (Total_Revolving_Bal), mengindikasikan bahwa semakin besar batas kredit yang belum digunakan, saldo berputar cenderung lebih rendah. Korelasi fitur jumlah tanggungan (Dependent_count) dengan fitur lain relatif lemah, termasuk korelasi negatif kecil dengan usia pelanggan. Heatmap ini membantu mengidentifikasi fitur-fitur yang saling terkait sehingga bisa menjadi pertimbangan penting dalam analisis atau pembuatan model prediksi.

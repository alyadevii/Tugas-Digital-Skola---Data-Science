Penggabungan sheet “Paid-Transaction” dan “Seller” di berkas excel “Homework-DataFrame-Dataset.xlsx”

```python
df_merge = pd.merge(paid_transaction, seller, on = "Order Number", how= "left")
df_merge
```

Pembuatan kolom baru bernama “Full Name” dan sisipkan kolom tersebut setelah kolom “Last
Name”

```python
df_merge ['Full Name'] = df_merge['First Name'] + ' ' + df_merge ['Last Name']
df_merge.insert (4, "Full Name", df_merge.pop("Full Name"))
df_merge
```

Pengubahan Nama lengkap menggunakan “title case”

```python
df_merge['Full Name']=df_merge['Full Name'].str.title()
df_merge
```

Pembuatan sebuah kolom setelah kolom “Seller Discount” dengan nama
“GMV”

```python
df_merge ['GMV'] = df_merge['Transaction Amount'] - df_merge['Seller Discount'] + df_merge['Delivery Fee']
df_merge.insert (9, "GMV", df_merge.pop("GMV"))
df_merge
```

Kelompokkan setiap Meta Category menjadi 2 group

```python
df_merge['Meta Category'].unique()
```

```python
category_group_mapping = {'Babies/ Kids':'Group 2',
                          'Service/ Mokado':'Group 2',
                          'Fashion':'Group 1',
                          'Gadget/ Komputer':'Group 2',
                          'Beauty/ Health':'Group 1',
                          'Home/ Food':'Group 1',
                          'Sports/ Hobi/ Otomotif':'Group 1',
                          'Elektronik':'Group 1'}
```

```python
df_merge ['Category'] = df_merge['Meta Category'].map(category_group_mapping)
df_merge.insert (6, "Category", df_merge.pop("Category"))
df_merge
```

Pembuatan DataFrame “Paid-Transaction” untuk menampilkan
jumlah total nilai GMV per-bulan untuk setiap group.

```python
df_merge['Month'] = df_merge['Paid Date'].dt.to_period('M')
pivot_gmv_per_month_per_group = pd.pivot_table(
    data= df_merge,
    index='Month',
    columns='Category',
    values='GMV',
    aggfunc='sum')
pivot_gmv_per_month_per_group
```

Python statement untuk mencari Seller mana yang memiliki total nilai GMV paling tinggi
di bulan Agustus 2017

```python
df_merge.info()
```

```python
df_merge['Paid Date'] = pd.to_datetime(df_merge['Paid Date'].astype(str), format='%Y%m%d')
df_merge
```

```python
df_merge.info()
```

```python
df_agustus = df_merge[df_merge['Paid Date']. dt.to_period('M') == '2017-08']
df_agustus
```

```python
df_gmv_sum_agustus = df_agustus.groupby(['Seller']).agg({'GMV': 'sum'})
df_gmv_sum_agustus.sort_values(by='GMV', ascending=False).head(1)
```

Python statement untuk mencari Count Seller banyaknya transaksi
di bulan September 2017 untuk Category “Fashion”.

```python
df_september = df_merge[df_merge['Paid Date']. dt.to_period('M') == '2017-09']
df_september
```

```python
df_september['Meta Category'].unique()
```

```python
fashion = df_september[df_september['Meta Category'] == 'Fashion']
```

```python
fashion_count = pd.pivot_table(data= fashion, index= 'Seller', values= 'Transaction Amount', aggfunc= 'count')
fashion_count.sort_values(by='Transaction Amount', ascending=False).head(1)
```
